#Seaborn library
Seaborn is a powerful Python library for creating statistical visualizations based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

```python
#1. Distribution plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = sns.load_dataset("tips")
sns.histplot(df["total_bill"], kde=True)
plt.title(" distribution of Total Bill ")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
#2.Box plot
sns.boxplot(x='day',y='total_bill',data=df)
plt.title('Total Bill by Day')
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#3.violin plot
sns.violinplot(x='day',y='total_bill',data=df)
plt.title('Total Bill by Day')
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#4.Bar plot
sns.barplot(x='day',y='total_bill',data=df,estimator=sum)
plt.title('Total Bill by Day')
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#5.scatter plot
sns.scatterplot(x='day',y='total_bill',data=df,hue='time')
plt.title('Total Bill vs Day')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
#6.lineplot
sns.lineplot(x='size',y='total_bill',data=df)
plt.title('Total Bill by size')
plt.xlabel("size")
plt.ylabel("Total Bill")
plt.show()
```

```python
#7.heatmap
import pandas as pd
tips_numeric=pd.get_dummies(df,drop_first=True)
c=tips_numeric.corr()
sns.heatmap(c,annot=True,cmap='rainbow')
plt.show()
```

```python
#8.pair plot
sns.pairplot(df,hue='time')
plt.show()
```

```python
#9.facet grid
g=sns.FacetGrid(df,col="time",row="day")
g.map(sns.histplot,"total_bill")
plt.show()
```

```python
#10 regression plot
import seaborn as sns
import matplotlib.pyplot as plt
#load a sample dataset
df = sns.load_dataset("tips")
#create a regression plot
sns.regplot(x='total_bill',y='tip',data=df)
plt.title("regression plot of total bill vs tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
#11.residual plot
sns.residplot(x='total_bill',y='tip',data=df,lowess=True)
plt.title("regression plot of total bill vs tip")
plt.xlabel("Total Bill")
plt.ylabel("Residual")
plt.show()
```
